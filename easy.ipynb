```python
def longestCommonPrefix(A):
    index = 0
    while True:
        cmp = [word[index] for word in A]
        if len(set(cmp)) == 1:
            index += 1
        else:
            break
    return A[0][:index] if index > 0 else ''

A = ["flower", "flow", "flight"]
#A = ["deio", "we", "ytq"]
print(longestCommonPrefix(A))
```
