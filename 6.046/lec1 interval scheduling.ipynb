
## Greedy : what count as best?

```python
start_L = [1,3,2,7,4]
end_L = [3,9,5,8,7]

# findLargest = lambda L: L.index(max(L))

def greedy(use, count):
    if len(use) == 0:
        return count
    
    # 1. find max
    maxIndex = use[0]
    maxValue = end_L[maxIndex]
    for i in use:
        if end_L[i] > maxValue:
            maxValue = end_L[i]
            maxIndex = i
    
    # 2. reset use
    newUse = use[:]
    newUse.remove(i)
    
    # 3. recursion
    count += 1
    return greedy(newUse, count)

use = [0,1,2,3,4]
count = 0
greedy(use, count)
```
