```python
def findOne(nums, s):
    if s == 0:
        return True
    elif len(nums) == 0:
        return False
    # recursive
    first = nums[0]
    rest = nums[1:]
    if first == s:
        return True
    if first < s:
        return findOne(rest, s-first) or findOne(rest, s)
    if first > s:
        return findOne(rest, s)

findOne([1,2,4],3)
#findOne([1,2,3,4],3)

    
```

```python
def helper(nums, k, target, start, curSum, visited):
    if k == 1:
        return True
    if curSum == target:
        return helper(nums, k-1, target, 0, 0, visited)
    for i in range(start, len(nums)):
        if visited[i]:
            continue
        visited[i] = True
        if helper(nums, k, target, i+1, curSum + nums[i], visited):
            return True
        visited[i] = False
    return False
        
```

```python
def canPartitionKSubsets(nums, k):
    if sum(nums) % k != 0:
        return False
    target = sum(nums) // k
    visited = [False] * len(nums)
    return helper(nums, k, target, 0, 0, visited)
```

## insertaion sort

```python

def insertSort(A):
    for j in range(1, len(A)):
        key = A[j]
        i = j - 1
        while i >= 0 and A[i] > key:
            A[i+1] = A[i]
            i -= 1
        A[i+1] = key
        
A = [3,2,6,0,8]
insertSort(A)
A
```

## quick sort


```python
def partition(A, p, r):
    x = A[r]
    i = p - 1
    j = p
    for j in range(p, r):
        if A[j] <= x:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i+1], A[r] = A[r], A[i+1]
    return i+1

def quickSort(A, begin, end):
    if begin < end:
        q = partition(A, begin, end)
        quickSort(A, begin, q-1)
        quickSort(A, q+1, end)
    return

A = [2,8,7,1,3,5,6,4]
quickSort(A, 0, len(A)-1)   
A
```

## random select : kth elem

```python
import random

def Rondom_partition(A, p, r):
    i = random.randint(p,r)
    A[r], A[i] = A[i], A[r]
    return partition(A, p, r)

def Random_select(A, p, r, i):
    if p == r:
        return A[p]
    q = Rondom_partition(A, p, r)
    k = q - p + 1
    if i <= k:
        return Random_select(A, p, q, i)
    else:
        return Random_select(A, q+1, r, i-k)

A = [2,8,7,1,3,5,6,4]
Random_select(A, 0, len(A)-1, 3)
```

## Dynamic prog

```python
def DP(nums):
    if len(nums) == 1:
        return True
    first = nums[0]
    
    ans = False
    for i in range(1, first+1):
        rest = nums[i:]
        # see if exceeds total length
        if i > len(rest):
            break
        if DP(rest):
            ans = True
            
    return ans

nums = [2,3,1,1,4]
nums = [3,2,1,0,4]
DP(nums)
```
